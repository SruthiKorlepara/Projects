```python
data.drop(["Cabin","Ticket","Name","Fare"],axis=1,inplace=True)
print(data)
```

```python
data["Age"].median()
```

```python
data["Age"]=data["Age"].fillna(value=27) # filling null values
data
```

```python
data.corr()
```

```python
plt.hist(data["Embarked"], bins=15)
plt.xlabel("Sex")
plt.ylabel("Embarked")
plt.show()
```

```python
sns.boxenplot(x=data["SibSp"], y=data["PassengerId"])

plt.xlabel("SibSp")
plt.ylabel("PassengerId")
plt.title("Passengers")
plt.show()
```

```python
class_counts = data["Survived"].value_counts()
plt.pie(class_counts.values, labels=class_counts.index, autopct='%1.1f%%')
plt.title("survived and not survived")
plt.show()
     
```

```python
sex_counts = data["Sex"].value_counts()
plt.bar(sex_counts.index, sex_counts.values,color="green")
plt.xlabel("Sex")
plt.xticks([0, 1])
plt.ylabel("Sibsp")
plt.title("Passenger counts by sex")
plt.show()
```

```python
sns.countplot(x=data['Parch'],hue=data['Embarked'])
plt.show()
```

```python
plt.boxplot(data["Age"])
plt.xlabel("Age")
plt.title("Ages of Passengers")
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True,fmt='.2%')
```

```python
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap="cool")
plt.title("Correlation matrix")
plt.show()
```
